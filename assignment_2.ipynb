In many specialized populations, there is little variability among the members. Is this a natural consequence of optimal selection? Here we generate a large number of populations from a given probabiliy density function, we pick the best members, and look at the distribution of the optimal selection. 

Import useful modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

Set the number of members in each population

```python
n_members=4
```

Generate a large number of populations in order to obtain a distribution of optimal members

```python
n_populations=100000
optimal_members_4=np.zeros(n_populations)
for i in np.arange(n_populations):
    generic_population=np.random.uniform(0,1,size=n_members)
    optimal_members_4[i]=np.max(generic_population)
```

We can plot the distribution of optimal members using the hist() function of matplotlib.pyplot

```python
plt.hist(optimal_members_4,histtype='step',density=True)
plt.show()
```

Numpy has the convenient functions np.average() and np.var() that allow to compute the average and variance of a collection of data

```python
print(np.average(optimal_members_4),np.var(optimal_members_4))
```

We can repeat the same procedure for a different number of members in the population. What happens to the distribution of optimal_members as the number of members in the population become large? What happen to its average and variance?

```python
n_members=10
n_populations=100000
optimal_members_10=np.zeros(n_populations)
for i in np.arange(n_populations):
    generic_population=np.random.triangular(0,0,1,size=n_members)
    optimal_members_10[i]=np.max(generic_population)
```

```python
plt.hist(optimal_members_4,histtype='step',density=True)
plt.hist(optimal_members_10,histtype='step',density=True)
plt.show()
```

```python
print(np.average(optimal_members_10),np.var(optimal_members_10))
```

```python
n_members=1000
n_populations=100000
optimal_members_100=np.zeros(n_populations)
for i in np.arange(n_populations):
    generic_population=np.random.triangular(0,0,1,size=n_members)
    optimal_members_100[i]=np.max(generic_population)
plt.hist(optimal_members_4,histtype='step',density=True)
plt.hist(optimal_members_10,histtype='step',density=True)
plt.hist(optimal_members_100,histtype='step',density=True)
plt.show()
```
